Todo parte de separar la ecuación de Schro... en cada sección:
$$
\frac{d^2}{dx^2} \psi + k_i^2 \psi = 0
$$

donde en la primera sección:
$$
k_1 = \sqrt{2mE/\hbar^2}
$$
en la segunda donde se encuentra el potencial presente:
$$
k_2= \sqrt{2m(E-V_0)/\hbar^2}
$$

# Caso 1: E > V_0
En este caso todos los $k_i \in \mathbb R$ por tanto las funciones de onda y su corriente de probabilidad $J$ no se ven afectadas

$$
\vec J_i = \frac{i\hbar}{2m} ( \psi_i \nabla \psi_i^* - \psi_i^* \nabla \psi_i )
$$

debido a que en este problema solo definimos corriente de probabilidad proviniente de la izquierda a la derecha, no existe corriente de probabilidad que provenga de la derecha a la izquierda, permitiendo eliminar un componente

utilizando las condiciones de contorno podemos obtener los valores de las constantes en función de la constante que corresponde a la onda incidente, la cual convenientemente se cancela en estos coeficientes:

asi los coeficientes de transmisión y reflexión:
$$
T = \frac{|J_{transmitido}|}{|J_{incidente}|} = \frac{4 k_1 k_2}{(k_1 + k_2)^2}
$$

$$
R = \frac{|J_{reflejado}|}{|J_{incidente}|} = \frac{(k_1 - k_2)^2}{(k_1 + k_2)^2}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

V0 = 5 # pongamosle  GeV de ejemplo, no sé si es mucho o poco
m = 1  # 1 me (1 masa electronica)
hbar = 1 # se cambian las unidades en que medimos la energia
Vmax = V0 + 5

# caso 1 E > V0
k1 = lambda E: np.sqrt( 2*m*E/hbar )
k2 = lambda E: np.sqrt( 2*m*(E-V0)/hbar )

E = np.linspace(V0, Vmax, 100)
```

```python
# definimos los coeficientes de relfexión y transmision
# como funciones
T = lambda k1, k2: 4*k1*k2/(k1+k2)**2
R = lambda k1, k2: (k1-k2)**2/(k1+k2)**2

# creamos una lista ejecutando estas funciones
T_array = T(k1(E), k2(E))
R_array = R(k1(E), k2(E))
```

```python
plt.plot(E, T_array, label=r"$T = \frac{ 4 k_1 k_2 }{(k_1 + k_2)^2 }$")
plt.plot(E, R_array, label=r"$R = \frac{(k_1 - k_2)^2}{(k_1 + k_2)^2}$")
plt.xlim(0, Vmax)
plt.legend()
plt.vlines(x=V0, ymin = 0, ymax = 1.0, alpha = 0.1, linestyles='dotted')
# debido a hbar = 1; me = 1, se debe de cambiar las unidades de la energia a esto:
plt.xlabel(r'$E \cdot \; 3.805 \times 10^{-20} \, [eV] $')
plt.ylabel('T,R')
plt.title(r'Salto Potencial en $V_0 = 5  \cdot \; 3.805 \times 10^{-20} \, [eV]$')
plt.show()
```
