# Aproximacion de Seno con Taylor

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)


plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, x , '--',label='taylor orden 1')
plt.plot(x, x - x**3/6 ,  '--',label='taylor orden 3')
plt.plot(x, x - x**3/6 + x**5/120,  '--',label='taylor orden 5')
plt.plot(x, x - x**3/6 + x**5/120 - x**7/(7*6*120), '--', label='taylor orden 7')
plt.legend()
plt.title(r'Serie Taylor $\sin x$')
plt.plot()
```

$$
\cos x = \sum_{n=0}^\infty  \frac{(-1)^n}{(2n)!} x^{2n}
$$

```python
plt.plot(x, np.cos(x), label='cos(x)')
plt.plot(x, 1 + 0*x , '--',label='taylor orden 0')
plt.plot(x, 1 - x**2/2 ,  '--',label='taylor orden 2')
plt.plot(x, 1 - x**2/2 + x**4/(4*3*2),  '--',label='taylor orden 4')
plt.plot(x, 1 - x**2/2 + x**4/(4*3*2) - x**6/(6*5*4*3*2) + x**8/(8*7*6*5*4*3*2), '--', label='taylor orden 8')
plt.legend()
plt.title(r'Serie Taylor $\cos x$')
plt.plot()
```
